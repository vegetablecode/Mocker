# Get file

```python
weather_data_location = './data/weather_data.csv'
col_names = ['station_id', 'station_name', 'year', 'month', 'day', 'avg temp C', 'tm status', 'avg humidity', 'hm status', 'wind speed', 'wd status', 'clouds', 'cd status']

weather_data = pd.read_csv(weather_data_location, names=col_names, sep=',', encoding='latin-1', header=None)

weather_data.head()
```

# Find location names

```python
station_names = weather_data.iloc[:,1]
station_names.head()
```

```python
station_names.unique()
```

# Vectors

```python
cities = ['BABIMOST']
station_data = weather_data[weather_data['station_name'].isin(cities)]
station_data.head()
```

# Print query

```python
id_device = 102
id_location = 41

hour = "12.00.00.000000 PM"

iterator = 1 # measure every *iterator* days

base = "INTO measurements (id_device, date_time, value, id_location) VALUES"
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# print sql query
print("INSERT ALL")
for index, row in station_data.iterrows():
    year = int(row['year'])
    month = months[int(row['month'])-1]
    day = int(row['day'])
    value = int(row['avg temp C'])
    
    print("{} ({}, '{}-{}-{} {}', {}, {})" .format(base, id_device, day, month, year, hour, value, id_location))
print("SELECT * FROM dual;")
```
